3. Find the most frequent value and their indice(s) in the following arrays:
i. x = np.array([1,2,3,4,5,1,2,1,1,1])
ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1,1,1,2,3,4,2,4,3,3])

def frequency(arr):
    values, counts = np.unique(arr, return_counts=True)

    max_count = np.max(counts)

    most_freq_values = values[counts == max_count]

    print("Array:", arr)
    print("Most frequent value(s):", most_freq_values)

    for v in most_freq_values:
        indices = np.where(arr == v)[0]
        print(f"Indices of {v}:", indices)
frequency(x)
frequency(y)
```

4. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),
